# Clustering Analysis after PCA 

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


# Select the number of components based on the scree plot
n_components = 100
df_pca = x_pca_df.iloc[:, :n_components]

# K-Means Clustering
kmeans = KMeans(n_clusters=5)  
clusters = kmeans.fit_predict(df_pca)

# t-SNE for Visualization
tsne = TSNE(n_components=2, perplexity=50, n_iter=300)
tsne_results = tsne.fit_transform(df_pca)

# Plot t-SNE
plt.figure(figsize=(16,10))
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=clusters, cmap='viridis', marker='o')
plt.title('t-SNE visualization of PCA components')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
import numpy as np

# Count the points in each cluster
unique, counts = np.unique(clusters, return_counts=True)
cluster_counts = dict(zip(unique, counts))

# Plot t-SNE
plt.figure(figsize=(16,10))
scatter = plt.scatter(tsne_results[:,0], tsne_results[:,1], c=clusters, cmap='viridis', marker='o')

# Calculate the centroids for each cluster in t-SNE space
centroids = np.array([tsne_results[clusters == i].mean(axis=0) for i in unique])

# Annotate the clusters with their counts
for i, centroid in enumerate(centroids):
    plt.text(centroid[0], centroid[1], str(cluster_counts[i]), weight='bold', ha='center', va='center', fontsize=12)

plt.title('t-SNE visualization of PCA components with Cluster Counts')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

# Heirarchical clustering 

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# Standardizing the data for hierarchical clustering
scaler = StandardScaler()
df_pca_scaled = scaler.fit_transform(df_pca)

# Performing hierarchical clustering
Z = linkage(df_pca_scaled, method='ward') # 'ward' is one of the methods

# Plotting the dendrogram
plt.figure(figsize=(16, 10))
dendrogram(Z, 
           truncate_mode='lastp',  
           p=12,                   
           leaf_rotation=45.,
           leaf_font_size=12.,
           show_contracted=True,   
)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
plt.axhline(y=50, c='k')   
plt.show()
```

# Identifying important genes through Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

# Load data (assuming 'df' is your DataFrame)
# df = pd.read_csv('your_data.csv')  # Uncomment if you need to load data from a file

# If 'cluster' is categorical, encode it
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
random_forest.fit(X_train, y_train)

# Get the feature importances from the model
importances = random_forest.feature_importances_

# Get the indices of the sorted features by importance
indices = np.argsort(importances)[::-1]

# Print the top 10 important features
print("Top 10 Important Features:")
for i in range(10):
    print(f"{X_train.columns[indices[i]]} (Importance: {importances[indices[i]]:.4f})")
```

```python
top_n = 10

# Get the names of the top important genes
top_genes = [df.columns[indices[i]] for i in range(top_n)]
```

# ANOVA

```python
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Assuming 'top_genes_data' is your DataFrame and includes 'cluster' and gene expression columns
# Ensure 'cluster' is treated as a categorical variable and the data is formatted correctly

top_genes_data['clusters'] = top_genes_data['clusters'].astype('category')  # Convert 'cluster' to a categorical type

# Perform ANOVA for each gene
for gene in top_genes:
    formula = f'{gene} ~ C(clusters)'
    model = ols(formula, data=top_genes_data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(f"ANOVA for {gene}:")
    print(anova_table)
    print("\n")
```

# SVM performance metrics

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame and 'clusters' is your target column
X = df.drop(columns='clusters')  # Features
y = df['clusters']  # Target

# Automatically encode all categorical variables to numeric
X = pd.get_dummies(X, drop_first=True)  # One-hot encoding

# Convert target variable 'clusters' to numeric if it's categorical
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Initialize and train the SVM
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X_train, y_train)

# Predict and evaluate the model
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)

# Output results
print("SVM Accuracy:", accuracy_svm)
print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm))

# Feature importances for the linear kernel
if svm.kernel == 'linear':
    importances_svm = svm.coef_[0]
    indices_svm = np.argsort(abs(importances_svm))[::-1]
    print("Top 10 Important Features:")
    for i in range(10):
        print(f"{X_train.columns[indices_svm[i]]} (Importance: {importances_svm[indices_svm[i]]:.4f})")

    # Plot the feature importances
    plt.figure(figsize=(10, 6))
    plt.title("Top 10 Feature Importances (SVM)")
    plt.bar(X_train.columns[indices_svm[:10]], importances_svm[indices_svm[:10]], color='blue')
    plt.xticks(rotation=45)
    plt.xlabel('Features')
    plt.ylabel('Importance')
    plt.tight_layout()
    plt.show()
```
